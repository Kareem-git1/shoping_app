# preprocessing


## split data

```python
x=df.drop('target',axis=1) # the input
y=df['target']       
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

### handle messing value & scaling the numerical columns

```python
df
```

```python
x.isna().mean()*100
```

```python
from sklearn.impute import SimpleImputer
num_imputer = SimpleImputer(strategy='most_frequent')

x_train[['is_active']] = num_imputer.fit_transform(x_train[['is_active']])
x_test[['is_active']]  = num_imputer.transform(x_test[['is_active']])
```

```python
cat_cols = ['gender','category']
cat_imputer = SimpleImputer(strategy='most_frequent')

x_train[cat_cols] = cat_imputer.fit_transform(x_train[cat_cols])
x_test[cat_cols]  = cat_imputer.transform(x_test[cat_cols])
```

```python
cat_imputer = SimpleImputer(strategy='median')

x_train[['signup_year']] = cat_imputer.fit_transform(x_train[['signup_year']])
x_test[['signup_year']]  = cat_imputer.transform(x_test[['signup_year']])
```

```python
x_train.isna().mean()*100
```

#### scaling

```python
from sklearn.preprocessing import RobustScaler
robust_scaler=RobustScaler()

num_col=x_train.select_dtypes(include='number').columns
x_train[num_col]=robust_scaler.fit_transform(x_train[num_col])
x_test[num_col]=robust_scaler.transform(x_test[num_col])
```

## encoding

```python
cat_col=x_train.select_dtypes(include='object').columns
cat_col
```

```python
for  col in cat_col:
    print(col)
    print(x_train[col].nunique())
    print(x_train[col].unique())
    print('-'*8)
    
```

```python
from sklearn.preprocessing import OrdinalEncoder


category_order = ['A','B','C','D']

ord_enc = OrdinalEncoder(categories=[category_order])

x_train[['category']] = ord_enc.fit_transform(x_train[['category']])
x_test[['category']] = ord_enc.transform(x_test[['category']])

```

##### encode nominal

```python
ohe_col_name=x_train.select_dtypes(object).columns
ohe_col_name
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop='first',sparse_output=False)

ohe_arr_train=ohe.fit_transform(x_train[ohe_col_name])
ohe_arr_test=ohe.transform(x_test[ohe_col_name])
```

```python
ohe_train=pd.DataFrame(ohe_arr_train,columns=ohe.get_feature_names_out()).reset_index()
ohe_test=pd.DataFrame(ohe_arr_test,columns=ohe.get_feature_names_out()).reset_index()
```

```python
x_train=x_train.reset_index()
x_test=x_test.reset_index()
y_test=y_test.reset_index()
y_test=y_test.reset_index()
```

```python
x_train=pd.concat([x_train,ohe_train],axis=1).drop(ohe_col_name,axis=1)
x_test=pd.concat([x_test,ohe_test],axis=1).drop(ohe_col_name,axis=1)
```
